```python
%pip install duckdb
import duckdb

conn = duckdb.connect()

conn.execute("""
    CREATE TABLE economic_data AS 
    SELECT * FROM read_csv_auto('../data/raw/world_bank_raw.csv')
""")

print("DuckDB big data engine initialized and raw data loaded successfully!")
```

```python
summary_df = conn.execute("""
    SELECT 
        ROUND(AVG(gdp_growth), 2) as avg_global_gdp_growth,
        ROUND(AVG(inflation), 2) as avg_global_inflation,
        ROUND(AVG(unemployment), 2) as avg_global_unemployment,
        COUNT(*) as total_years_countries_processed
    FROM economic_data
""").df()

summary_df
```
